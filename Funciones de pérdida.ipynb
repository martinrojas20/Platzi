MSE - Error Cuadrat√≠co Medio

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mse(y, y_hat, derivate=False):
  if derivate:
    return y_hat - y
  else:
    return np.mean((y_hat - y)**2)
```

```python
real = np.array([0,1,1,0,1,1,0])
prediction = np.array([0.87, 0.1, 0.65, 0.92, 0.02, 0.05, 0.87])
```

```python
mse(real, prediction)
```
