```python
numeros = []
for num in range(0, 10000):
  num = str(num)
  if len(num) == 1:
    num = '000'+ num
    numeros.append(num)
  elif len(num) == 2:
    num = '00' + num
    numeros.append(num)
  elif len(num) == 3:
    num = '0' + num
    numeros.append(num)
  else:
    numeros.append(num)

  
```

```python
dic = {'numeros': numeros}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dic)
```

```python
df.info()
```

```python
df.to_csv("numeros.csv", index=False, encoding='utf-8', sep=',')
```
