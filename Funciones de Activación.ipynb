```python
import numpy as np
import matplotlib.pyplot as plt
```

Funci√≥n Sigmoide

```python
def sigmoide(x):
  return 1 / (1 + np.exp(-x))
```

```python
num = np.linspace(10, -10, 100)
```

```python
plt.plot(num)
```

```python
res = sigmoide(num)
```

```python
plt.plot(res)
```

```python
plt.plot(num, sigmoide(num))
```

Funci√≥n RELU

```python
def relu(val):
  return np.maximum(0, val)
```

```python
plt.plot(num, relu(num))
```

Funci√≥n Tangente

```python
def tanh(val):
  return (np.exp(val ) - np.exp(-val)) / ((np.exp(val) + np.exp(-val)))
```

```python
valores = np.linspace(-5,5,100)
```

```python
plt.plot(valores, tanh(valores))
```
